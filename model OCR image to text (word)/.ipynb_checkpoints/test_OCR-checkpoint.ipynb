```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
# load le modèle sauvegardé
model_path = "OCR_model.keras"
model = tf.keras.models.load_model(model_path)

# load class_names
class_names = np.load('class_names.npy')
```

```python
def predict_word_from_image(image_path, model, class_names):
    """
    Segments letters from an image and predicts the word using a pre-trained model.

    Parameters:
        image_path (str): Path to the input image.
        model: Pre-trained model to predict letters.
        class_names (list): List of class names corresponding to the model's outputs.

    Returns:
        str: Predicted word.
    """
    # Load the image
    image = cv2.imread(image_path)
    if image is None:
        raise ValueError("Image not found at the provided path.")

    # Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply thresholding for better segmentation
    _, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

    # Find contours
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Sort contours from left to right based on x-coordinate
    contours = sorted(contours, key=lambda x: cv2.boundingRect(x)[0])

    # Extract segmented letters
    segmented_letters = []
    for contour in contours:
        x, y, w, h = cv2.boundingRect(contour)
        letter_roi = image[y:y + h, x:x + w]  # Extract the letter ROI
        segmented_letters.append(letter_roi)

    # Predict letters and form the word
    predicted_word = ""
    for letter_roi in segmented_letters:
        # Preprocess the letter ROI
        letter = cv2.resize(letter_roi, (64, 64))
        letter = np.array(letter, dtype=np.float32) / 255.0
        letter = letter.reshape(1, 64, 64, 3)  # Reshape to match model input shape

        # Predict the letter using the pre-trained model
        predictions = model.predict(letter)
        predicted_label_index = np.argmax(predictions, axis=1)

        # Get the predicted label (character)
        predicted_label = class_names[predicted_label_index[0]]

        # Append the predicted letter to the word
        predicted_word += predicted_label

    # Display the original image
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title("Input Image")
    plt.axis("off")
    plt.show()

    return predicted_word
```

```python
# Example usage (requires a model and class names):
predicted_word = predict_word_from_image('test1.png', model, class_names)
print("Predicted Word:", predicted_word)
```
